```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('intfloat/multilingual-e5-small' )

documents = [
    "Старинный замок возвышался на скале; его стены помнят осады и рыцарские турниры.",
    "Он проверил, хорошо ли закрыт замок на входной двери, и повернул ключ два раза."
]

query = "как правильно запереть дверь?"

doc_embeddings = model.encode(documents, convert_to_tensor=True)
query_embedding = model.encode(query, convert_to_tensor=True)

similarities = util.pytorch_cos_sim(query_embedding, doc_embeddings)[0]


print(f"Запрос: {query}\n")
for i, doc in enumerate(documents):
    print(f"Документ {i+1}:")
    print(f"  Текст: {doc}")
    print(f"  Косинусное сходство с запросом: {similarities[i]:.4f}\n")

best_idx = similarities.argmax().item()
print(f"Наиболее релевантный документ: {best_idx+1} (сходство = {similarities[best_idx]:.4f})")
```
